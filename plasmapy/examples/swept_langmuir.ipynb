```python
import astropy.units as u
import numpy as np
import os
import sys
import xarray as xr

sys.path.insert(0, os.path.abspath('../../'))

from plasmapy.diagnostics import xdiagnostics
```

### Enable the Swept Langmuir Diagnostic

By default `plasmapy` does not register any of its `xarray.Dataset` accessor to avoid polution of the namespace and reduce overhead of imported code.  Instead, we provide the `xdiagnostic` object so the user can enable the `xarray.Dataset` accessor(s) he or she wants.  Just executing `xdiagnostics` will show what diagnosics are available and enabled/disabled.

```python
xdiagnostics.enable('swept_langmuir')
xdiagnostics
```

### Build the `xarray.Dataset` for Swept Langmuir Analysis

```python
# Load the bias and current values stored in the .p pickle file
path = os.path.join('./', 'langmuir_samples', "Beckers2017.npy")
bias, current = np.load(path)

# voltage DataArray
da_volt = xr.DataArray(bias,
                       dims=['time_index'],
                       name='voltage')

# current DataArray
da_curr = xr.DataArray(current,
                       dims=['time_index'],
                       name='current')

# build dataset
ds = xr.Dataset({da_volt.name: da_volt,
                 da_curr.name: da_curr})

# set probe parameters
# ds.plasmapy.swept_langmuir.probe_parameters = lp

ds
```
