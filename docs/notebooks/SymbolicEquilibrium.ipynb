# `SymbolicEquilibrium` and figures of merit

```python
from plasmapy.plasma.symbolicequilibrium import SymbolicEquilibrium

params = {"aspect_ratio": 0.32, "A": -0.155, "elongation": 1.7, "triangularity": 0.33}
eq = SymbolicEquilibrium(**params, B0=5.2, config="non-null")
eq.plot();
```

```python
tau = np.linspace(0, 2 * np.pi, 300)
x = 1 + eq.aspect_ratio * np.cos(tau + np.arcsin(eq.triangularity) * np.sin(tau))
y = eq.aspect_ratio * eq.elongation * np.sin(tau);

ax = eq.plot()
flux_surface = eq.get_flux_surface(0)
flux_surface.plot(ax, n=True);
ax.plot(x, y)
```
