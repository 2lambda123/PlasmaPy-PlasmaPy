## `SymbolicEquilibrium`

```python
from plasmapy.plasma.symbolicequilibrium import SymbolicEquilibrium

params = {"aspect_ratio": 0.32, "A": -0.155, "elongation": 1.7, "triangularity": 0.33}
eq = SymbolicEquilibrium(**params, B0=5.2, config="single-null")
eq.plot();
```

## `FluxSurface`

```python
ax = eq.plot()
fs = flux_surface = eq.get_flux_surface(-0.01)
fs.plot(ax, n=True);
centroid = np.array([np.mean(fs.R), np.mean(fs.Z)])
plt.plot(*centroid, "ro", label="Centroid")
plt.legend()
```

```python
ax = eq.plot()
fs.plot(ax, n=True);
centroid = np.array([np.mean(fs.R), np.mean(fs.Z)])
plt.plot(*centroid, "ro", label="Centroid")
plt.plot(circle_x + centroid[0], circle_y + centroid[1], label = "Equivalent circle")
plt.legend()
```
